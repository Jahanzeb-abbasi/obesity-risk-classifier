```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/raw/train.csv")
```

```python
print(df.shape)
print(df.info())
```

```python
df.isnull().sum()
```

```python
df["NObeyesdad"].value_counts()
```

```python
#Splitting columns into categorical & numerical lists

categorial_cols = df.select_dtypes(include=["object", "string"]).columns.tolist()
categorial_cols.remove("NObeyesdad")

numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns.tolist()
numeric_cols.remove("id")

print("Categorical features:", categorial_cols)
print("Numerical features:", numeric_cols)
```

```python
print(df.CAEC.value_counts())
print(df.CALC.value_counts())
```

```python
from sklearn.preprocessing import StandardScaler

#Mapping integers in Ordinal Columns
freq_map = {"no": 0, "Sometimes": 1, "Frequently": 2, "Always": 3}
df["CAEC"] = df["CAEC"].map(freq_map)
df["CALC"] = df["CALC"].map(freq_map)
```

```python
nominal_cols = [c for c in categorial_cols if c not in ["CAEC", "CALC"]]
print(nominal_cols)
df = pd.get_dummies(df, columns=nominal_cols, drop_first=True)
```

```python
df
```

```python
#Scaling numerics for Logistic Regression
scaler = StandardScaler()
df_scaled_numeric = scaler.fit_transform(df[numeric_cols])
```

```python
print(df.head())
print(df.shape)
```

```python
y = df["NObeyesdad"]

feature_cols = [c for c in df.columns if c not in ["id", "NObeyesdad"]]

#For Random Forest - unScaled numeric values
X_rf = df[feature_cols]

#For Logistic Regression - Scaled numeric version
X_logreg = X_rf.copy()
X_logreg[numeric_cols] = df_scaled_numeric
```

```python
print(X_rf.shape)
print(X_logreg.shape)
```

```python
from sklearn.model_selection import train_test_split

X_rf_train, X_rf_test, X_logreg_train, X_logreg_test, y_train, y_test = train_test_split(
    X_rf, X_logreg, y,
    test_size = 0.2,
    random_state=42,
    stratify=y
)
```

```python
print(X_logreg_train.shape)
print(X_logreg_test.shape)
```

```python
print("TRAIN",y_train.value_counts(normalize=True))
print("TEST",y_test.value_counts(normalize=True))
```

```python
import joblib

joblib.dump({
    "X_rf_train": X_rf_train,
    "X_rf_test": X_rf_test,
    "X_logreg_train": X_logreg_train,
    "X_logreg_test": X_logreg_test,
    "y_train": y_train,
    "y_test": y_test,
    "scaler": scaler,
    "feature_cols": feature_cols,
}, "../artifacts/preprocessed_data.joblib")
```
